```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\sahan\Downloads\hapiscore_whr.xlsx")
df.set_index("country", inplace=True)
df = df.transpose()
df
```

```python
df = df.iloc[:, :10]
df
```

```python
plt.figure(figsize=(10, 3))
for country in df.columns:
    plt.plot(df.index, df[country], marker="o", linestyle="--", linewidth=1, label=country)
plt.ylabel("Happiness Score", fontsize=12)  # Corrected y-axis label
plt.xlabel("Years", fontsize=12)
plt.title("Happiness Over Time (First 10 Countries)", fontsize=14, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(loc="best", fontsize=4)
plt.tight_layout()
plt.show()
```
