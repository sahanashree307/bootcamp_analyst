```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r"C:\Users\sahan\Downloads\life expectancy.xlsx")
df.set_index("country",inplace=True)
df=df.transpose()
```

```python
df
```

```python
df["Average"]=df.mean(axis=1)
```

```python
plt.figure(figsize=(5,5))
```

```python
plt.plot(df["Average"],df.index,marker="o",linestyle="--",color="Blue",linewidth=1,label="Average")
plt.ylabel("Year",fontsize=12)
plt.xlabel("Average life expectancy(Years)",fontsize=12)
plt.title("Life expectancy trends over time",fontsize=14,fontweight="bold")
plt.grid(True,linestyle="--",alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
